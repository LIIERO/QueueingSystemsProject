```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df['Time spent in queue [s]'] = df['Time spent in queue [s]'].astype(float)
df['Time spent in system [s]'] = df['Time spent in system [s]'].astype(float)
df['Priority'] = df['Priority'].astype(str)
df['Class name'] = df['Class name'].astype(str)


plt.figure(figsize=(8, 6))
sns.scatterplot(x="Time spent in queue [s]", y="Time spent in system [s]", hue="Priority", style="Class name", s=100, palette="cool", data=df)
plt.title("Czas w systemie vs czas w kolejce (z podziałem na priorytety)")
plt.xlabel("Czas spędzony w kolejce [s]")
plt.ylabel("Czas spędzony w systemie [s]")
plt.legend(title="Priorytet")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure that columns are properly formatted
df['ID'] = df['ID'].astype(int)
df['Priority'] = df['Priority'].astype(str)

# Create the line plot with priorities only
plt.figure(figsize=(12, 8))  # Increased figure size for better readability
sns.lineplot(x="ID", y="Time spent in system [s]", hue="Priority", markers=True, dashes=False, palette="viridis", data=df, marker='o')

# Add titles and labels
plt.title("Czas spędzony w systemie na przestrzeni ID pacjentów", fontsize=16)
plt.xlabel("ID pacjenta", fontsize=14)
plt.ylabel("Czas spędzony w systemie [s]", fontsize=14)

# Improve legend
plt.legend(title="Priorytet", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)

# Customize grid and ticks
plt.grid(True)
plt.xticks(rotation=45)
plt.yticks(fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Histogram of Time Spent in System
plt.subplot(1, 2, 1)
sns.histplot(df['Time spent in system [s]'], kde=True, bins=30, color='blue')
plt.title('Rozkład czasu spędzonego w systemie')
plt.xlabel('Czas spędzony w systemie [s]')
plt.ylabel('Liczba pacjentów')

# Histogram of Time Spent in Queue
plt.subplot(1, 2, 2)
sns.histplot(df['Time spent in queue [s]'], kde=True, bins=30, color='green')
plt.title('Rozkład czasu spędzonego w kolejce')
plt.xlabel('Czas spędzony w kolejce [s]')
plt.ylabel('Liczba pacjentów')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Box Plot by Priority
plt.subplot(1, 2, 1)
sns.boxplot(x='Priority', y='Time spent in system [s]', data=df,order=order, palette="Set3")
plt.title('Czas spędzony w systemie według priorytetu')
plt.xlabel('Priorytet')
plt.ylabel('Czas spędzony w systemie [s]')

# Box Plot by Class
plt.subplot(1, 2, 2)
sns.boxplot(x='Class name', y='Time spent in system [s]', data=df,order=order_age, palette="Set3")
plt.title('Czas spędzony w systemie według klasy')
plt.xlabel('Klasa')
plt.ylabel('Czas spędzony w systemie [s]')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Był w tylu systemach', y='Time spent in system [s]', hue='Priority', style='Class name', s=100, palette="cool", data=df)
plt.title('Czas spędzony w systemie vs liczba systemów')
plt.xlabel('Liczba systemów')
plt.ylabel('Czas spędzony w systemie [s]')
plt.legend(title='Priorytet')

max_systems = df['Był w tylu systemach'].max()
plt.xticks(range(3, max_systems + 2, 1))
plt.show()
```

```python
plt.figure(figsize=(12, 6))
heatmap_data = df.pivot_table(values='Time spent in system [s]', index='Class name', columns='Priority', aggfunc='mean')
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="YlGnBu", cbar_kws={'label': 'Średni czas spędzony w systemie [s]'})
plt.title('Średni czas spędzony w systemie według klasy i priorytetu')
plt.xlabel('Priorytet')
plt.ylabel('Klasa')
plt.show()
```

```python

import numpy as np

# Assuming df is your DataFrame
# Convert 'Time [s]' to float
df['Time [s]'] = pd.to_numeric(df['Time [s]'], errors='coerce')

# Drop rows with NaN values that couldn't be converted
df = df.dropna(subset=['Time [s]'])

# Determine the number of bins or set specific bin edges
bins = np.arange(0, df['Time [s]'].max() + 0.5, 0.5)

# Group by time intervals of 0.5s
df_grouped = df.groupby(pd.cut(df["Time [s]"], bins=bins)).mean()

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df_grouped.iloc[:, 1:], cmap="coolwarm", annot=True, fmt=".2f", cbar_kws={'label': 'Średnia liczba pacjentów'})
plt.title("Średnia liczba pacjentów w kolejkach w interwałach czasowych")
plt.xlabel("Kolejki")
plt.ylabel("Interwały czasowe")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
for column in df.columns[1:]:
    sns.histplot(df[column], kde=True, bins=10, label=column, alpha=0.5)
plt.title("Histogram liczby pacjentów w kolejkach")
plt.xlabel("Liczba pacjentów")
plt.ylabel("Częstotliwość")
plt.legend(title="Kolejki")
plt.grid(True)
plt.show()
```
