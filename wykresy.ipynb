```python
import numpy as np
from matplotlib.ticker import MultipleLocator

# Filtracja: porównujemy obecny, poprzedni i następny punkt
filtered_data = df_que.copy()

# Usuwanie skoków większych o 1 względem sąsiednich wartości
for column in df_que.columns[1:]:
    for i in range(1, len(df_que) - 1):
        prev_val = df_que[column].iloc[i - 1]
        curr_val = df_que[column].iloc[i]
        next_val = df_que[column].iloc[i + 1]
        
        # Jeśli obecna wartość jest większa o 1 niż obie sąsiednie, wygładzamy
        if abs(curr_val - prev_val) >= 1 and abs(curr_val - next_val) >= 1:
            filtered_data[column].iloc[i] = prev_val  # Średnia z poprzedniej i następnej wartości

# Przygotowanie danych do wykresu skumulowanego
time = filtered_data["Time [s]"]
queues = filtered_data.columns[1:]  # Pomijamy kolumnę "Time [s]"
data = [filtered_data[column] for column in queues]

# Rysowanie wykresu skumulowanego
plt.figure(figsize=(14, 8))
plt.stackplot(time, data, labels=queues, alpha=0.8, edgecolor='black')

# Dodanie pionowej linii na 260 sekundzie
plt.axvline(x=260, color='blue', linestyle='--', linewidth=2, label='Zamknięcie rejestracji')

# Ustawienia osi
plt.gca().xaxis.set_major_locator(MultipleLocator(50))  # Oś X z odstępami co 50 jednostek
plt.title("Liczba pacjentów w kolejkach w czasie (skumulowany, po filtracji)")
plt.xlabel("Czas [s]")
plt.ylabel("Liczba pacjentów")
plt.legend(title="Kolejki", loc="upper left", bbox_to_anchor=(1, 1))  # Legenda poza wykresem
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()  # Automatyczne dopasowanie elementów
plt.show()
```

```python

import numpy as np

# Assuming df is your DataFrame
# Convert 'Time [s]' to float
df['Time [s]'] = pd.to_numeric(df['Time [s]'], errors='coerce')

# Drop rows with NaN values that couldn't be converted
df = df.dropna(subset=['Time [s]'])

# Determine the number of bins or set specific bin edges
bins = np.arange(0, df['Time [s]'].max() + 0.5, 0.5)

# Group by time intervals of 0.5s
df_grouped = df.groupby(pd.cut(df["Time [s]"], bins=bins)).mean()

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df_grouped.iloc[:, 1:], cmap="coolwarm", annot=True, fmt=".2f", cbar_kws={'label': 'Średnia liczba pacjentów'})
plt.title("Średnia liczba pacjentów w kolejkach w interwałach czasowych")
plt.xlabel("Kolejki")
plt.ylabel("Interwały czasowe")
plt.show()
```
